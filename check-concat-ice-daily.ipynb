````python
import xarray as xr
import glob
````

````python
# %%time
ok = True
# basedir = '/g/data3/hh5/tmp/cosima/access-om2-01/01deg_jra55v13_iaf/output001/ice/OUTPUT/'
# basedir = '/scratch/x77/aek156/access-om2/archive/01deg_jra55v140_iaf/output243/ice/OUTPUT/'
# basedir = '/g/data/ik11/outputs/access-om2-01/01deg_jra55v140_iaf/output000/ice/OUTPUT/'
basedir = '/scratch/x77/aek156/access-om2/archive/01deg_jra55v140_iaf_cycle2/output244/ice/OUTPUT/'
cdailies = glob.glob(basedir+'iceh.????-??-daily.nc')
cdailies.sort()
for cdaily in cdailies:
    print(cdaily)
    ds_concat = xr.open_dataset(cdaily)
#     dailies = glob.glob(cdaily.replace('-daily', '-??'))
#     dailies = glob.glob(cdaily.replace('-daily', '-??').replace('.nc', '.nc-DELETE'))
    dailies = glob.glob(cdaily.replace('-daily', '-??').replace('.nc', '.nc-DEPRECATED')) # for output244
    dailies.sort()
    for d, daily in enumerate(dailies):
        print(daily)
        ds_day = xr.open_dataset(daily)
        # don't check the following
        ds_day.attrs.update({'comment2': ds_concat.comment2})
        ds_day.attrs.update({'history': ds_concat.history})
        ds_day.attrs.update({'NCO': ds_concat.NCO})
#         ds_day.attrs.update({'nco_openmp_thread_number': ds_concat.nco_openmp_thread_number})
#         print(ds_day.isel(time=0).identical(ds_concat.isel(time=d)))
#         print(ds_day.isel(time=0).equals(ds_concat.isel(time=d)))
        c = ds_day.isel(time=0)
        t = ds_concat.isel(time=d)
        identical = c.identical(t)
        if not identical:
            equal = c.equals(t)
        else:
            equal = True  # identical implies equal
        if equal and not identical:
            print(c.attrs)
            print(t.attrs)
        if not (equal and identical):
            ok = False
    #         print(ctrlFile)
    #         print(testFile)
            print(equal, identical)
            cvars = list(c.keys())
            cvars.sort()
            tvars = list(t.keys())
            tvars.sort()
            for cvar, tvar in zip(cvars, tvars):
                idcl = c[cvar].identical(t[tvar])
                if not idcl:
                    eql = c[cvar].equals(t[tvar])
                else:
                    eql = True  # identical implies equal
                print('    ', cvar, tvar, eql, idcl)
    #         print()
        else:
            print('ok')
        print('')
if ok:
    print('All ok')
````
