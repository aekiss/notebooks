Check the CICE grid coordinates

```python
import xarray as xr, numpy as np
```

```python
om2cice1_grid = xr.open_dataset('/g/data/ik11/inputs/access-om2/input_20201102/cice_1deg/grid.nc')
om2mom1_grid = xr.open_dataset('/g/data/ik11/inputs/access-om2/input_20201102/mom_1deg/ocean_hgrid.nc')
```

```python
# Comparison with MOM supergrid
print("Coordinate errors for 1 degree grid")
diff = np.rad2deg(om2cice1_grid.tlat.values) - om2mom1_grid.y[1::2,1::2]
print('TLAT % 10.3e' % abs(diff).max().values)
diff = np.rad2deg(om2cice1_grid.tlon.values) - om2mom1_grid.x[1::2,1::2]
print('TLON % 10.3e' % abs(diff).max().values)

# NE points for U
diff = np.rad2deg(om2cice1_grid.ulat.values) - om2mom1_grid.y[2::2,2::2]
print('ULAT % 10.3e' % abs(diff).max().values)
diff = np.rad2deg(om2cice1_grid.ulon.values) - om2mom1_grid.x[2::2,2::2]
print('ULON % 10.3e' % abs(diff).max().values)
```
